```python
df = df.round({'alpha': 5, 'delta': 5}) # url for getting image is in format of 5 decimal places
```

```python
import requests
import os

# Built in API for SDSS 1 --> https://skyserver.sdss.org/dr1/en/help/docs/api.asp
# No API for SDSS17 --> https://skyserver.sdss.org/dr17/VisualTools/quickobj

def fetch_sdss_image(object_id, output_directory='images'):
    os.makedirs(output_directory, exist_ok=True)

    url = f"https://skyserver.sdss.org/dr17/SkyServerWS/ImgCutout/getjpeg?TaskName=Skyserver.Chart.Navi&ra={object_id['ra']}&dec={object_id['dec']}&scale=0.099031675&radius=0.2&format=json"
    response = requests.get(url)

    if response.status_code == 200:
        image_path = os.path.join(output_directory, f"{object_id['id']}.jpg")
        with open(image_path, 'wb') as img_file:
            img_file.write(response.content)
        print(f"Image saved to {image_path}")
    else:
        print(f"Failed to retrieve image: {response.status_code}")
```

```python
for id in range(100):
  object_id_info = {
    'id': df["obj_ID"][id],  # Replace with your object ID
    'ra': df["alpha"][id],    # Replace with the object's right ascension
    'dec': df["delta"][id]   # Replace with the object's declination
  }
  fetch_sdss_image(object_id_info)
```
